## Build Dataset

```python
from torch.utils.data import Dataset, DataLoader

class TempDataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        inputs = self.x[idx] # [5, 7]
        labels = self.y[idx]
        return inputs, labels
```

## Build Model

```python
import torch
import torch.nn as nn

class TempPredictionModel(nn.Module):
    def __init__(self, num_features, hidden_size, hidden_size_2, output_size, num_layers):
        super().__init__()
        self.lstm = nn.LSTM(num_features, hidden_size, num_layers, batch_first=True, dropout=0.5)

        self.fc1 = nn.Linear(hidden_size, hidden_size_2)

        self.relu = nn.ReLU()

        self.fc2 = nn.Linear(hidden_size_2, output_size)

    def forward(self, x):
        # x: [batch size, seq len, num_features]
        out, _ = self.lstm(x)
        # out: [batch size, seq len, hidden size]
        out = out[:, -1, :]
        # out: [batch size, hidden size]
        out = self.relu(self.fc1(out))
        # out: [batch size, output size]
        out = self.fc2(out)
        return out

```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)

import random

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

seed = 42
set_seed(seed)

num_features = 7
hidden_size = 50
hidden_size_2 = 20
output_size = 1
num_layers = 2
model = TempPredictionModel(num_features, hidden_size, hidden_size_2, output_size, num_layers)
```
